### 3. Analysis Using the Natural Language Toolkit (NLTK)

#### 3.1. Scraping and parsing the data

```python
import requests
from bs4 import BeautifulSoup
url = "https://blog.playstation.com/?smcid=pdc%3Aen-us%3Aprimary%20nav%3Amsg-news%3Aps-blog"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
text1 = soup.get_text()
```

#### 3.2. Tokenization and POS tagging

```python
import nltk
from nltk.probability import FreqDist
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

english_stop_words = stopwords.words('english')
english_stop_words.append('latam')  

text1 = text1.lower() 
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text1)
tokens = [token for token in tokens if token not in english_stop_words]
print(tokens)
```

```python
from collections import Counter
tags = nltk.pos_tag(tokens)
count = Counter( tag for word,  tag in tags)
print(count)
nltk.pos_tag(tokens)
```

#### 3.3. Word frequency

```python
fdist = FreqDist(tokens)
common = fdist.most_common(40)
fdist.most_common(40)
```

```python
import matplotlib.pyplot as plt
fdist.plot(40,cumulative=False, color = 'lightcoral', title = 'Most common words - PlayStation Blog')
plt.show()
```

```python
import matplotlib.pyplot as plt
words = [('playstation', 38), ('date', 19), ("2021", 19), ('published', 18), ('video', 17), ('ps4', 16), ('play', 14), ('trailer', 14), ('june', 13), ('1', 13), ('2', 11), ('may', 11), ('ps5', 10), ('ps', 9), ('plus', 9), ('duration', 8), ('content', 7), ('store', 7), ('director', 7), ('new', 7), ('blog', 6), ('vr', 6)] 
word_dict = {}
for w in words:
    word_dict[w[0]]=w[1]

plt.bar(range(len(word_dict)), word_dict.values(), align='center', color = 'lightcoral')
plt.xticks(range(len(word_dict)), word_dict.keys(), rotation = 90)
plt.ylabel('Frequency')
plt.title('Most common words - PlayStation Blog')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize=(30, 30))
wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='white', colormap='Pastel2', collocations=False, stopwords = STOPWORDS).generate(text1)
plt.axis("off")
plt.imshow(wordcloud)
plt.title("Relative importance of words", fontsize = 100, color = 'lightcoral')

plt.show()
```

#### 3.4 Collocations

```python
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(tokens)
finder.apply_freq_filter(4)
finder.nbest(bigram_measures.pmi, 20) 
```

Singular or mass nouns (NN) were recognized as a part of speech with the highest frequency. The high number of occurrences of singular or mass nouns can be seen as indicative of the heightened inclusion of platform-specific terms; for instance, among the common words were PlayStation-specific terms such as *PS4*, *PS*, *PS5*, *PlayStation*, *VR*, and *Sony* assigned to the NN tag. Adjectives (JJ), placed as the second most common POS, occurred three times less frequently than singular or mass nouns. The high frequency of cardinal numbers (CN) can be attributed to the layout of *PlayStation Blog* in which articles are displayed alongside the date of their publication. What is more, dates frequently occur within the textual dimension of gaming news due to the emphasis put on informing potential clients (gamers) about the upcoming game releases and release updates.

#### 4.3. Collocations
N-grams available in Python's *NLTK* package can serve as a tool of context analysis. According to Mohler, 'two words is the perfect number for capturing the key phrases and themes that provide context for entity sentiment'  ([2019](https://www.lexalytics.com/lexablog/context-analysis-nlp)). Therefore, bigrams were chosen to identify words that co-occur in text data. In order to ignore bigrams that occur less than four times, filtering was applied to eliminate weaker collocations. Due to the use of bigrams, the tokens pointed to in subsection 4.2. can be more efficiently represented as words associated with the gaming industry. For instance, the word *launch* collocates with *trailer*. Within the video game industry, the phrase launch trailer refers to the promotional material presenting the gameplay and storyline value of a given game usually presented at the trade event for the video game industry, *E3*. Tokens such as *VR*, *store*, and *plus* are further analyzed as tokens explicitly related to the platform-specific terminology. The *PlayStation VR* is described as a virtual reality headset developed by *Sony Interactive Entertainment* and a piece of gaming equipment compatible with home consoles. While the phrase *PlayStation Plus* denotes Sony's subscription service that enables online play for both PS4 and PS5 video games, *PlayStation Store* stands for Sony's online store service where one can download game titles and content. Additionally, the word *video* is explicitly linked to game trailers, particularly those released for the video game titles available for PS4 and PS5 consoles.
